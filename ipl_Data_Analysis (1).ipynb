```python
df1["won_by"]=df1["won_by"].map({"Wickets":"Chasing","Runs":"Defending"})
```

```python
px.pie(labels=label,values=counts,names=label)
```

```python
px.bar(x=label,y=counts)
```

```python
bowlers_count=df1["best_bowling"].value_counts()
bowlers_count=bowlers_count.head(5)
```

```python
max_score1 = df1.groupby("venue")["first_ings_score"].agg(np.max)
max_score1
```

```python
labels=max_score1.index
```

```python
px.bar(x=labels,y=max_score1,title="Highest run Score in first inning with Stadium")
```
